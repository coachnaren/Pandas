#Merging and Joining Data:

##Concatenating DataFrames:

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2'], 'B': ['B0', 'B1', 'B2']})
df2 = pd.DataFrame({'A': ['A3', 'A4', 'A5'], 'B': ['B3', 'B4', 'B5']})
result = pd.concat([df1, df2])
result
```

##Merging DataFrames:


```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2'], 'A': ['A0', 'A1', 'A2']})
right = pd.DataFrame({'key': ['K0', 'K1', 'K2'], 'B': ['B0', 'B1', 'B2']})
result = pd.merge(left, right, on='key')
result
```

#Aggregation and Grouping:


GroupBy operations.

Aggregation functions (sum, mean, count, etc.).

```python
import pandas as pd
iplsixdict={
    'Player':['Jos Buttler','Liam Livingstone','Andre Russell','Sanju Samson','KL Rahul',
              'Shimron Hetmyer','Dinesh Karthik','Suryakumar Yadav','Ambati Rayudu','Tilak Varma',
              'Nitish Rana','Shivam Dube','Robin Uthappa','Rahul Tripathi','Aiden Markram'],
    'Team':['RR','PBK','KKR','RR','LSG','RR','RCB','MI','CSK',"MI",'KKR','CSK','CSK','SRH','SRH'],
    'sixes':[36,23,22,21,20,19,17,15,15,15,15,14,14,13,12]
}
iplsixesDF=pd.DataFrame(iplsixdict)
iplsixesDF
```

```python
iplsixdict={
    'Player':['Jos Buttler','Liam Livingstone','Andre Russell','Sanju Samson','KL Rahul',
              'Shimron Hetmyer','Dinesh Karthik','Suryakumar Yadav','Ambati Rayudu','Tilak Varma',
              'Nitish Rana','Shivam Dube','Robin Uthappa','Rahul Tripathi','Aiden Markram'],
    'Team':['RR','PBK','KKR','RR','LSG','RR','RCB','MI','CSK',"MI",'KKR','CSK','CSK','SRH','SRH'],
    'sixes':[36,23,22,21,20,19,17,15,15,15,15,14,14,13,12]
}
iplsixesDF=pd.DataFrame(iplsixdict,index=range(1,16))
iplsixesDF
```

```python
iplsixesDF['Team'].unique()
```

```python
iplsixesDF['Team'].nunique()
```

```python
#groupby method
groupDF=iplsixesDF.groupby('Team')
groupDF
```

```python
groupDF.sum()
```

```python
# Grouping by Team and calculating the total sixes
total_sixes_by_team = iplsixesDF.groupby('Team')['sixes'].sum()
print(total_sixes_by_team)
```

#average six per player by team

```python
average_sixes_by_team = iplsixesDF.groupby('Team')['sixes'].mean()
print("Average Sixes per Player by Team:")
print(average_sixes_by_team)
```

```python
top_player_by_team = iplsixesDF.loc[iplsixesDF.groupby('Team')['sixes'].idxmax()]
print("Top Player in Each Team Based on Sixes:")
print(top_player_by_team[['Team', 'Player', 'sixes']])
```

```python
top_5_players = iplsixesDF.nlargest(5, 'sixes')
print("Top 5 Players with Most Sixes:")
print(top_5_players[['Player', 'Team', 'sixes']])
```

```python
total_sixes_by_player = iplsixesDF[['Player', 'sixes']].set_index('Player')
print("Total Sixes by Players:")
print(total_sixes_by_player)
```

```python
players_above_15_sixes = iplsixesDF[iplsixesDF['sixes'] > 15]
print("Teams with Players Hitting More Than 15 Sixes:")
print(players_above_15_sixes[['Team', 'Player', 'sixes']])
```

```python
count_players = iplsixesDF.groupby('Team')['Player'].count()
print("\nCount of Players by Team:")
print(count_players)
```
